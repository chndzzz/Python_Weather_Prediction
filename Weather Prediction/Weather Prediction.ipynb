```python
df_1 = df_final.set_index("Daily Summary")                     # Syntax to set the "Daily Summary" as index column
df_1
```

```python
# Dropping "Formatted Date" & "Precip Type" columns for further analysis of making visualization
df_2 = df_1.drop(['Formatted Date','Precip Type'],axis=1)
df_2
```

```python
# Grouping the DataFrame according to "Daily Summary" based on the mean value of other columns
df_grouped = df_2.groupby("Daily Summary").mean()
df_grouped
```

```python
# creating a function named "create_plot" that can be used in the interact widget
def create_plot(col_name):
  fig = make_subplots(rows=1,cols=3)                      # syntax to make subplot with 3 charts in 1 row

  for coln in df_grouped.columns.to_list():               # creating a for loop to select the coumn name from the DataFrame
    if col_name == coln:
      fig.add_trace(go.Bar(x=df_grouped.index,y=df_grouped[col_name],name=col_name),row=1,col=1)       # using graph_objects library we create Bar graph


  for coln in df_grouped.columns.to_list():               # creating a for loop to select the coumn name from the DataFrame
    if col_name == coln:  
      fig.add_trace(go.Line(x=df_grouped.index,y=df_grouped[col_name],name=col_name),row=1,col=2)       # using graph_objects library we create Line graph

  for coln in df_grouped.columns.to_list():               # creating a for loop to select the coumn name from the DataFrame
    if col_name == coln:  
      fig.add_trace(go.Box(x=df_grouped.index,y=df_grouped[col_name],name=col_name),row=1,col=3)       # using graph_objects library we create Boxplot

      fig.update_layout(
          width=2500,
          height=1200,
          autosize=True,
          margin=dict(t=0, b=0, l=0, r=0),
          template="plotly")

      fig.update_xaxes(title="Daily Summary")               # adding title to x-axis
      fig.update_yaxes(title=coln)                          # adding title to y-axis
      fig.show()
```

```python
# syntax to run the interact function and changing the graphs automatically
widgets.interact(create_plot,col_name=df_grouped.columns.to_list())
```

```python
# importing the datetime library to work on the date & time data
from datetime import datetime
from datetime import timezone
```

```python
# syntax to change the format of the date column 
df_final['Formatted Date']=pd.to_datetime(df_final['Formatted Date'],utc=True)
df_final['Formatted Date']
```

```python
# checking the changes in Formatted Data into the Dataframe
df_final
```

```python
# Syntax to extract the year from the Formatted Date column
df_year=df_final['Formatted Date'].dt.year
df_year
```

```python
# Syntax to drop the Formatted Date column
df_dropdate = df_final.drop(['Formatted Date'],axis=1)
df_dropdate
```

```python
# Syntax to concat the df_year Dataframe and df_dropdate Dataframe into 1 Dataframe
concate = [df_year,df_dropdate]
df_concat = pd.concat(concate,axis=1)
df_concat
```

```python
# Dropping the unnecessary columns from the Dataframe
df_date = df_concat.drop(['Temperature (C)','Humidity','Wind Speed (km/h)','Wind Bearing (degrees)','Visibility (km)','Pressure (millibars)','Daily Summary'],axis=1)
df_date
```

```python
# We need the data for "rain", so we need to remove the data which has "snow" as Precip Type
df_date.drop(df_date[(df_date['Precip Type']=='snow')].index,inplace=True)
df_date
```

```python
df_date['Precip Type'].unique()                  # now we can see Precip Type has only "rain" data
```

```python
# now groupby "Formatted Date" based on the count of Precip Type
df_dategrp = df_date.groupby('Formatted Date')['Precip Type'].count()             # this will give result in Series format
df_dategrp
```

```python
# converting the above Series into DataFrame and reseting the index number
df_bar=pd.DataFrame(df_dategrp)
df_bar.reset_index(inplace=True)
df_bar
```

```python
# Plotting the bar graph between number of times the particular year has got rain
plt.rcParams["figure.figsize"] = (15,7)                              # increases the size of graph
bar_plot=df_bar.plot.bar(x='Formatted Date',y='Precip Type',color=['Magenta','Green','Blue'],width=0.8,edgecolor='Black')
plt.title("Year with number of times it rained",fontdict={'fontsize':35,'color':'Red','fontweight':25})      # Formatting the title
bar_plot.tick_params(axis='x', colors='purple')                                                 # Formatting the x-axis values
bar_plot.tick_params(axis='y', colors='purple')                                                 # Formatting the y-axis values
plt.xlabel('Year',fontdict={'fontsize':25,'color':'Blue','fontweight':15})
plt.ylabel('Count of rain',fontdict={'fontsize':25,'color':'Blue','fontweight':15})
bar_plot.legend_=None                                                                           # Removing the legend from the visual
plt.show()
```
